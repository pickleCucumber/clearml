```python
import time

import pandas as pd
import requests
```

```python
url = "http://localhost:8080/serve/logreg"
```

```python
df = pd.read_csv("/home/dev/data.csv", index_col=0)
```

```python
df = df[['app_date', 'birthdate', 'gender', 'Score', 'not_app', 'typeemployerid', 'total_app', 'Goods_type', 'reject_app', 'request_amount', 'NBKI_RetailScore', 'income', 'nameTO', 'SiteName', '60_6mob']]
```

```python
df
```

```python
df.info(verbose=True, show_counts=True)
```

```python
while True:
    for index, row in df.iterrows():
        # Process each row to convert timestamps to strings and replace NaNs
        row_json = (
            row.apply(lambda x: x.isoformat() if isinstance(x, pd.Timestamp) else x)
            .where(pd.notnull(row), None)
            .to_dict()
        )
       # print(pd.DataFrame([row_json]))

        # Send the processed row as part of the payload
        payload = {
            "row": row_json,
        }
        response = requests.post(url, json=payload)
        time.sleep(0.3)

        print(response.text, row["60_6mob"])
```
